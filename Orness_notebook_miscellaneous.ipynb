# Hello World to the Deep learning

````python
# Importing the libraries for deep learning
import tensorflow as tf
import numpy as np
from tensorflow import keras
#from tensorflow.keras import layers
````

## Building our first Deep learning model

### Defining the model

Simplest NN possible:   
For Predicting the function ```y = 2x-1```

![linear_input](https://drive.google.com/uc?id=1CReCvkvxDonpN42TDuIPeShsKP8pKfua)

### Defining the architecture

````python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
````

### Compiling

We know that in our function, the relationship between the numbers is y=2x-1.  

For compiling you need to know 2 things:  
Optimizer: The function making guess
loss: calculating the distance from the true function  

And communicate back to optimizer. . .


````python
model.compile(optimizer='sgd', loss='mean_squared_error')
````

````python
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
````

### Fitting

````python
model.fit(xs, ys, epochs=10)
````

### Predicting

````python
print(model.predict([10.0]))
````

````python
model.summary()
````

## Testing on Wine quality

````python
import pandas as pd
path = "/content/drive/MyDrive/Colab_Notebooks/orness/clean_data/"
df = pd.read_csv(path+"winequality.csv")
````

````python
X = X = df.drop('TARGET', axis=1)
Y = df['TARGET'].apply(lambda y: 1 if y > 5 else 0)
````

````python
model = keras.Sequential([
    keras.layers.Dense(128, activation='relu', input_shape=[X.shape[1]]),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(1),
])
````

````python
model.compile(
    optimizer="adam", 
    loss="mae",
    metrics=['accuracy']
)
````

````python
model.fit(X, Y, epochs=25)
````

````python
input = np.asarray(df.drop('TARGET', axis=1).iloc[50]).reshape(1, -1)
model.predict([input])
````
