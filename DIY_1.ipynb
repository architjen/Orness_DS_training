```python
import pandas as pd
```

```python
df_1 = {'name': ['Archit', 'Etienne'],
        'surname':['JAIN', 'EKPO'],
        'id': [ 23, '29']
}
```

```python
#convert the dictionary df_1 to dataframe
df_1 = pd.DataFrame(df_1)
df_1
```

```python
df_2 = {'id': ['22', '47'],
        'surname': ['TCYSIN', 'KUNTA'],
        'name':['Alex', 'Dan']
}
```

```python
# convert the dictionary to dataframe
df_2 = pd.DataFrame(df_2)
```

```python
# concatenate the data (use pandas), make sure their columns are in order (?)
df = pd.concat([df_1, df_2], ignore_index= True)
df
```

```python
# print df and see the dtypes 
df 
```

```python
# clean the data set by correcting the columns types
# HINT: astype()
df['id'] = df['id'].astype(int)
df['name'] = df['name'].astype(str)
```
